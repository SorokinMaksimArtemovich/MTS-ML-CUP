```python
import pandas as pd
import numpy as np
import pickle
```

```python
def load(path, verbose=True):
    if verbose:
        print("Loading object from {}".format(path))
    with open(path, "rb") as obj_file:
        obj = pickle.load(obj_file)
    if verbose:
        print("Object loaded from {}".format(path))
    return obj
```

## In this notebook I average predictions from DaNet model on 3 datasets
First I averege only models with 10 folds and 100 epohs, but when I add models with other folds and epohes it improved result on 0,0003

```python
age = pd.read_csv('data/danet_10_50/df_age/age_predicts.csv')
age_1 = pd.read_csv('data/danet_10_50/df_age_1/age_predicts.csv')
danet_age = pd.read_csv('data/danet_10_50/df_danet_age/age_predicts.csv')
sex = pd.read_csv('data/danet_10_50/df_sex/is_male_predicts.csv')
sex_1 = pd.read_csv('data/danet_10_50/df_sex_1/is_male_predicts.csv')
sex_5_100 = pd.read_csv('data/danet_5_100/is_male_predicts.csv')
age_5_1 = pd.read_csv('data/danet_5_50/age_predicts (1).csv')
age_5_2 = pd.read_csv('data/danet_5_50/age_predicts (2).csv')
age_5_0 = pd.read_csv('data/danet_5_50/age_predicts.csv')
sex_5_0 = pd.read_csv('data/danet_5_50/is_male_predicts.csv')
sex_5_1 = pd.read_csv('data/danet_5_50/is_male_predicts (1).csv')
sex_5_2 = pd.read_csv('data/danet_5_50/is_male_predicts (1).csv')
age_cb_1 = pd.read_csv('data/catboost/submission_age_1.csv')
age_cb_0 = pd.read_csv('data/catboost/submission_age.csv')
cb_danet = pd.read_csv('data/catboost/submission_danet.csv')
sex_cb_1 = pd.read_csv('data/catboost/submission_sex_1.csv')
sex_cb_0 = pd.read_csv('data/catboost/submission_sex.csv')
```

```python
age = age.merge(age_1, on='user_id', suffixes=('', '_1'))
age = age.merge(danet_age, on='user_id', suffixes=('', '_2'))
age = age.merge(age_5_1, on='user_id', suffixes=('', '_3'))
age = age.merge(age_5_2, on='user_id', suffixes=('', '_4'))
age = age.merge(age_5_0, on='user_id', suffixes=('', '_5'))
age = age.merge(age_cb_1, on='user_id', suffixes=('', '_6'))
age = age.merge(age_cb_0[['user_id', 'age']], on='user_id', suffixes=('', '_7'))
age = age.merge(cb_danet[['user_id', 'age']], on='user_id', suffixes=('', '_8'))
age
```

```python
sex = sex.merge(sex_1, on='user_id', suffixes=('', '_1'))
sex = sex.merge(sex_5_100, on='user_id', suffixes=('', '_2'))
sex = sex.merge(sex_5_0, on='user_id', suffixes=('', '_3'))
sex = sex.merge(sex_5_1, on='user_id', suffixes=('', '_4'))
sex = sex.merge(sex_5_2, on='user_id', suffixes=('', '_5'))
sex = sex.merge(sex_cb_1, on='user_id', suffixes=('', '_6'))
sex = sex.merge(sex_cb_0, on='user_id', suffixes=('', '_7'))
sex = sex.merge(cb_danet[['user_id', 'is_male']], on='user_id', suffixes=('', '_8'))
sex
```

```python
sex['is_male'] = sex[['is_male', 'is_male_1', 'is_male_2', 'is_male_3', 'is_male_4', 'is_male_5', 'is_male_6', 'is_male_7', 'is_male_8']].mean(axis=1)
sex = sex.drop(['is_male_1', 'is_male_2', 'is_male_3', 'is_male_4', 'is_male_5', 'is_male_6', 'is_male_7', 'is_male_8'], axis=1)
sex
```

```python
age['age'] = age[['age', 'age_1', 'age_2', 'age_3', 'age_4', 'age_5', 'age_6', 'age_7', 'age_8']].mean(axis=1)
age = age.drop(['age_1', 'age_2', 'age_3', 'age_4', 'age_5', 'age_6', 'age_7', 'age_8'], axis=1)
age
```

```python
age['age'] = round(age['age']).astype(int)
age
```

```python
age = age.merge(sex, on='user_id')
age
```

```python
age.to_csv('mean_pred_danet&cb_all.csv', index=False)
```
